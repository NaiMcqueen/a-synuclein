```python
# Abrir y leer el archivo
with open('/content/rms_DHA_C2.txt', 'r') as file:
    lines = file.readlines()

# Crear una lista para almacenar las líneas que cumplen el criterio
filtered_lines = []
extracted_texts = []

for line in lines:
    # Extraer el valor de RMS después del ":"
    rms_value = float(line.split(':')[-1].strip())

    # Verificar si el RMS es menor a 2
    if rms_value < 2:
        filtered_lines.append(line)

        # Extraer la parte del texto entre "and" y ":"
        if "and" in line and ":" in line:
            extracted_text = line.split("and")[1].split(":")[0].strip()
            extracted_texts.append(extracted_text)

# Eliminar duplicados de los textos extraídos
unique_texts = list(set(extracted_texts))

# Crear la línea de texto con 'delete' seguido de los textos separados por comas
delete_command = "delete " + " and ".join(unique_texts)

# Escribir el comando en un nuevo archivo
with open('conformeros_delete_DHA_C2.txt', 'w') as output_file:
    output_file.write(delete_command)
```
